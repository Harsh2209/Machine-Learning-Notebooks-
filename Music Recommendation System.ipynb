## Music Recommendation System

The following code suggests Artists/Bands based on the artist you listen to(here the name of artist/band is taken as input).

The algorithm used is non-negative matrix factorization(NMF).

This project does not provide accuracy as the problems belongs to unsupervised learning category.

```python
from sklearn.preprocessing import StandardScaler

scale=StandardScaler()

scale.fit_transform(final)
```

```python
nmf=NMF(n_components=11)

nmf.fit(final)
features=nmf.transform(final)

features
```

```python
from sklearn.preprocessing import normalize

feat=normalize(features)
dff=pd.DataFrame(feat,index=final.index)
ab=dff.loc['Simon & Garfunkel']
sim=dff.dot(ab)

sim.nlargest(10)
```
