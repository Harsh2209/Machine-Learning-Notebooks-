# Wikipedia Pages Recommendation

```python
import pandas as pd
import numpy as np
```

```python
wiki=pd.read_csv('./MSIT_ML_CLASS-master/datasets/wikipedia/wikipedia-vectors.csv')
wiki.head()
```

```python
wiki.drop('Unnamed: 0',axis=1,inplace=True)
wiki.head()
```

```python
from pprint import pprint
print('List of columns : ')
pprint(wiki.columns)
```

```python
wiki.shape
```

```python
wiki=wiki.T
```

```python
wiki.shape
```

```python
wiki.head()
```

```python
from sklearn.decomposition import NMF

nmf=NMF(n_components=10)

nmf.fit(wiki)
features=nmf.transform(wiki)

features.shape
```

```python
features
```

```python
titles=list(wiki.index)
titles
```

```python
from sklearn.preprocessing import normalize

n_features=normalize(features)

df=pd.DataFrame(n_features,index=wiki.index)

ab=df.loc['Adam Levine']
sim=df.dot(ab)

print(sim.nlargest())
```

## Wikpedia vocabulary (words in the wikipedia pages)

```python
words=pd.read_csv('./MSIT_ML_CLASS-master/datasets/wikipedia/wikipedia-vocabulary-utf8.txt',names=['Words in Vocabulary'])
words.head()
```

```python
df=pd.DataFrame(nmf.components_,columns=words)
df.head()
```

```python
nmf.components_
```

```python
for i in range(10) :
    comp=df.iloc[i]
    print(comp.nlargest())
    print('\n')
```
